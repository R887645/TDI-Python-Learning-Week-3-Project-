```python
#Task1:#create a list of all order costs, Find the total of all orders,the average order value, 
#Sort the list and identify the median value 

import csv

# Step 1: Load dataset
total_cost = []

with open("supply_chain.csv", "r") as file:
    reader = csv.DictReader(file)
    for row in reader:
        # Use the exact column name from your CSV header
        total_cost.append(float(row['total_cost']))

# Step 2: Total of all orders
total_orders = sum(total_cost)

# Step 3: Average order value
average_order = total_orders / len(total_cost)

# Step 4: Sort the list
total_cost.sort()

# Step 5: Median
n = len(total_cost)
if n % 2 == 0:
    median_order = (total_cost[n//2 - 1] + total_cost[n//2]) / 2
else:
    median_order = total_cost[n//2]

# Print results
print("List of Total Costs:", total_cost)
print("Total of All Orders:", total_orders)
print("Average Order Value:", average_order)
print("Median Order Value:", median_order)
```

```python
#Task2: Map Products to Suppliers Using Dictionaries

import csv

# Step 1: Initialize an empty dictionary to store product-supplier mapping
product_supplier = {}

# Step 2: Open the dataset file (CSV format)
with open("supply_chain.csv", "r") as file:
    reader = csv.DictReader(file)  # Read rows as dictionaries
    
    # Step 3: Loop through each row in the dataset
    for row in reader:
        product = row['product_category']   # Extract product name
        supplier = row['supplier_name']     # Extract supplier name
        
        # Step 4: Build dictionary mapping products to suppliers
        # If product already exists in dictionary, append supplier (avoid duplicates)
        if product in product_supplier:
            if supplier not in product_supplier[product]:
                product_supplier[product].append(supplier)
        else:
            # If product not yet in dictionary, create a new entry with supplier in a list
            product_supplier[product] = [supplier]

# Step 5: Look up which supplier(s) provide a specific product
product_to_lookup = "Motors"   # Change this to test other products
suppliers_for_product = product_supplier.get(product_to_lookup, "Product not found")

# Step 6: Count how many unique suppliers exist across all products
all_suppliers = set()  # Use a set to avoid duplicates
for suppliers in product_supplier.values():
    all_suppliers.update(suppliers)
unique_supplier_count = len(all_suppliers)

# Step 7: Print results for clarity
print("Product-Supplier Dictionary:", product_supplier)
print(f"Suppliers for {product_to_lookup}:", suppliers_for_product)
print("Number of Unique Suppliers:", unique_supplier_count)
```

```python
#Task3:Find Unique Values with Sets
import csv

# Step 1: Initialize empty sets for suppliers, products, and order statuses
unique_suppliers = set()
unique_products = set()
unique_statuses = set()

# Step 2: Also build a dictionary mapping products to suppliers (to check uniqueness later)
product_supplier_map = {}

# Step 3: Open and read the dataset
with open("supply_chain.csv", "r") as file:
    reader = csv.DictReader(file)
    
    for row in reader:
        supplier = row['supplier_name']
        product = row['product_category']
        status = row['delivery_status']
        
        # Add values to sets (automatically keeps only unique items)
        unique_suppliers.add(supplier)
        unique_products.add(product)
        unique_statuses.add(status)
        
        # Build product-to-supplier mapping
        if product in product_supplier_map:
            product_supplier_map[product].add(supplier)
        else:
            product_supplier_map[product] = {supplier}

# Step 4: Print each set
print("Unique Suppliers:", unique_suppliers)
print("Unique Products:", unique_products)
print("Unique Order Statuses:", unique_statuses)

# Step 5: Count how many unique items exist in each category
print("Number of Unique Suppliers:", len(unique_suppliers))
print("Number of Unique Products:", len(unique_products))
print("Number of Unique Order Statuses:", len(unique_statuses))

# Step 6: Find products that only one supplier provides
products_with_one_supplier = [product for product, suppliers in product_supplier_map.items() if len(suppliers) == 1]

print("Products with only one supplier:", products_with_one_supplier)
```

```python
#task 4 : Identify Single-Source Products

# Assume you already built product_supplier_map in Task 2
# Example structure: {'Motors': ['SupplierB', 'SupplierA'], 'Valves': ['SupplierC']}

# Step 1: Loop through the product-to-supplier dictionary
print("=== Supply Chain Risk Report: Single-Source Products ===")

for product, suppliers in product_supplier_map.items():
    # Step 2: Identify products with only one supplier
    if len(suppliers) == 1:
        # Step 3: Print a risk report entry
        print(f"⚠️ RISK: Product '{product}' is supplied only by {suppliers[0]}")
```

```python
#task 5: Build a Complex Network

import csv

# Step 1: Initialize an empty dictionary to store supplier -> products mapping
supplier_network = {}

# Step 2: Read the dataset
with open("supply_chain.csv", "r") as file:
    reader = csv.DictReader(file)
    
    for row in reader:
        supplier = row['supplier_name']         # Supplier name
        product = row['product_category']       # Product name
        cost = float(row['total_cost'])         # Total cost for that order
        
        # Step 3: Build nested structure
        # Each supplier maps to a list of dictionaries (products + costs)
        if supplier not in supplier_network:
            supplier_network[supplier] = []
        
        supplier_network[supplier].append({
            "product": product,
            "cost": cost
        })

# Step 4: Query the structure (example: show all products from Supplier A)
supplier_to_lookup = "Supplier A"
if supplier_to_lookup in supplier_network:
    print(f"Products from {supplier_to_lookup}:")
    for item in supplier_network[supplier_to_lookup]:
        print(f" - {item['product']} (Cost: {item['cost']})")
else:
    print(f"{supplier_to_lookup} not found in dataset.")

# Step 5: Calculate total value of products from each supplier
print("\n=== Total Value by Supplier ===")
for supplier, items in supplier_network.items():
    total_value = sum(item['cost'] for item in items)
    print(f"{supplier}: {total_value}")
```

```python
#Extra Challenges
#challenge 1 : Orders Exceeding a Cost Threshold
#Create a list of orders that exceed a certain cost threshold. Then calculate what percentage 
#of total orders this represents. 

import csv

# Step 1: Load all order costs into a list
total_costs = []

with open("supply_chain.csv", "r") as file:
    reader = csv.DictReader(file)
    for row in reader:
        total_costs.append(float(row['total_cost']))

# Step 2: Define a threshold (example: 20,000)
threshold = 20000

# Step 3: Create a list of orders that exceed the threshold
orders_above_threshold = [cost for cost in total_costs if cost > threshold]

# Step 4: Calculate percentage of total orders this represents
percentage_above_threshold = (len(orders_above_threshold) / len(total_costs)) * 100

# Step 5: Print results
print("Orders Above Threshold:", orders_above_threshold)
print(f"Number of Orders Above {threshold}: {len(orders_above_threshold)}")
print("Total Number of Orders:", len(total_costs))
print(f"Percentage of Orders Above {threshold}: {percentage_above_threshold:.2f}%")


```

```python
#Challenge 2: 
#Using your supplier-to-products dictionary, find which supplier provides the most products.

import pandas as pd

# ==== 1) Load the CSV ====
csv_path = "supply_chain.csv"
df = pd.read_csv(csv_path)  # add sep=";" if needed, e.g., pd.read_csv(csv_path, sep=";")

# Normalize headers (lowercase, strip) to avoid subtle mismatches
df.columns = [c.strip().lower() for c in df.columns]

# ==== 2) Identify the actual column names in your file ====
# From your screenshot the key columns are likely:
#   - supplier_name
#   - product_c  (appears to be the product / product category column)
candidate_product_cols = ["product_c", "product", "product_category", "product_name", "product_type"]
product_col = next((c for c in candidate_product_cols if c in df.columns), None)

if product_col is None:
    raise KeyError(
        "Could not find a product column. Looked for: "
        f"{candidate_product_cols}. Found columns: {list(df.columns)}"
    )

supplier_col = "supplier_name"
if supplier_col not in df.columns:
    raise KeyError(
        f"Expected supplier column '{supplier_col}' not found. "
        f"Available columns: {list(df.columns)}"
    )

# (Optional) clean values (trim spaces)
df[product_col] = df[product_col].astype(str).str.strip()
df[supplier_col] = df[supplier_col].astype(str).str.strip()

# ==== 3) Build supplier -> set(products) mapping ====
supplier_products = {}
for _, row in df[[supplier_col, product_col]].dropna().iterrows():
    supplier = row[supplier_col]
    product = row[product_col]
    if supplier in supplier_products:
        supplier_products[supplier].add(product)
    else:
        supplier_products[supplier] = {product}

# ==== 4) Find which supplier provides the most distinct products ====
max_supplier = None
max_count = 0

for supplier, products in supplier_products.items():
    if len(products) > max_count:     # fixed: '>' not '&gt;'
        max_count = len(products)
        max_supplier = supplier

# ==== 5) Print results ====
print("=== Supplier Product Coverage Report ===")
if max_supplier is None:
    print("No suppliers/products found.")
else:
    print(f"Supplier with the most products: {max_supplier}")
    print(f"Number of products supplied: {max_count}")
    print(f"Products supplied: {sorted(supplier_products[max_supplier])}")
```

```python
#challenge3
#Create a structure that groups orders by status. Then query it to find all delivered orders 
#and calculate their average value.

import pandas as pd
from collections import defaultdict

# === 1) Load CSV ===
csv_path = "supply_chain.csv"  # update if needed
df = pd.read_csv(csv_path)

# Normalize column names to avoid header mismatches
df.columns = [c.strip().lower() for c in df.columns]

# Validate required columns present exactly as in your file
required_cols = {"delivery_status", "total_cost"}
missing = required_cols - set(df.columns)
if missing:
    raise KeyError(
        f"Missing required column(s): {missing}. "
        f"Available columns: {list(df.columns)}"
    )

# Ensure numeric total_cost
df["total_cost"] = pd.to_numeric(df["total_cost"], errors="coerce")

# === 2) Create a structure that groups orders by status ===
# We'll build: dict[str, list[dict]]   e.g., groups["Delivered"] -> list of row dicts
groups = defaultdict(list)
for _, row in df.iterrows():
    status = str(row["delivery_status"]).strip()
    groups[status].append(row.to_dict())

# (Optional) quick overview using pandas groupby
summary = df.groupby("delivery_status", dropna=False).agg(
    orders_count=("delivery_status", "size"),
    total_value=("total_cost", "sum"),
    avg_value=("total_cost", "mean")
).reset_index()
print("=== Summary by delivery_status ===")
print(summary.to_string(index=False))

# === 3) Query the structure for all Delivered orders ===
delivered_key = "Delivered"  # matches the case as shown in your file
delivered_orders = groups.get(delivered_key, [])

print(f"\nDelivered orders found: {len(delivered_orders)}")

# === 4) Calculate average value of Delivered orders ===
# Extract total_cost safely and compute mean
delivered_total_costs = [
    float(order["total_cost"]) for order in delivered_orders
    if order.get("total_cost") is not None and str(order["total_cost"]).strip() != "" and pd.notna(order["total_cost"])
]

avg_delivered_value = sum(delivered_total_costs) / len(delivered_total_costs) if delivered_total_costs else float("nan")

print(f"Average delivered order value: {avg_delivered_value:,.2f}")

# (Optional) Save the delivered orders to a CSV if you want
# pd.DataFrame(delivered_orders).to_csv("delivered_orders_only.csv", index=False)
```
